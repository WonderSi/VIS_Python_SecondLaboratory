Условие
Дана строка, в которой буква h встречается как минимум два раза. Разверните
последовательность символов, заключенную между первым и последним появлением буквы
h , в противоположном порядке.

```python
string = str(input())
h_positions = []
index = 0

while index < len(string):
    if string[index] == 'h':
        h_positions.append(index)
    index += 1

if len(h_positions) >= 2:
    result = ''
    i = 0
    while i < len(h_positions) - 1:
        first_h, last_h = h_positions[i], h_positions[i + 1]
        reversed_part = string[first_h + 1:last_h][::-1]
        result += reversed_part
        i += 2
    print(result)
else:
    print('В строке меньше 2-ух h')
```

Изобразите извилистый спуск улитки по стене по заданной траектории.
Вводится одна строка. Начальный символ этой строки — символ рисования пути улитки.
Далее следует последовательность символов “<”, “>” и “V”, которые означают движение
улитки, соответственно, влево, вправо и вниз.
Гарантируется, что по любой горизонтали движение происходит только в одну сторону
(иными словами, между идущими в любом порядке “<” и “>” всегда есть хотя бы один “V”)
и что путь не пройдёт левее начального положения.
Выводятся несколько строк, состоящих из символов рисования пути и пробелов. Верхний
левый угол соответствует начальному положению улитки. Каждый последующий символ
входной строки означает, что нужно дорисовать символ рисования пути левее, правее или
ниже предыдущего. Пробелы используются для отступов.

```python
try:
    path = str(input())
    if not path:
        raise ValueError('Путь не может быть пустым!')
    x, y = 0, 0
    positions = {}
    drawing_char = path[0]
    positions[(x, y)] = drawing_char

    for char in path[1:]:
        if char == '<':
            x -= 1
        elif char == '>':
            x += 1
        elif char == 'V':
            y += 1
        positions[(x, y)] = drawing_char

    min_x = min(pos[0] for pos in positions)
    max_x = max(pos[0] for pos in positions)
    min_y = min(pos[1] for pos in positions)
    max_y = max(pos[1] for pos in positions)

    for y in range(min_y, max_y + 1):
        row = ''
        for x in range(min_x, max_x + 1):
            row += positions.get((x, y), ' ')
        print(row)
except ValueError as e:
    print(f'Ошибка: {e} Повторите ввод!')
```

Давайте попробуем на самом деле растянуть слова, как будто они резиновые. Возьмём за
середину и будем тянуть. Если в слове нечётное количество букв, на верхушке окажется
одна буква. А вот если чётное, то тянуть будем за две средние буквы, они и окажутся
наверху. Напишите программу, которая покажет, как будет выглядеть растянутое слово.
При решении задачи можно использовать только циклы и операции над строками и
числами

```python
try:
    string = str(input("Введите слово: "))  ###############################################################
    length_string = len(string)

    if length_string == 0:
        raise ValueError('Слово не может быть пустым!')

    if length_string % 2 == 1:
        # Нечётное количество букв
        middle = length_string // 2
        top = string[middle]  # Средняя буква
        left = string[:middle]  # Левая часть
        right = string[middle + 1:]  # Правая часть
    else:
        # Чётное количество букв
        middle = length_string // 2
        top = string[middle - 1:middle + 1]  # Две средние буквы
        left = string[:middle - 1]  # Левая часть
        right = string[middle + 1:]  # Правая часть

    # Выводим верхнюю часть
    print(" " * len(left) + top)

    # Выводим растянутые части
    for i in range(len(left)):
        print(left[len(left) - 1 - i] + " " * (2 * i + 1) + right[i])

except ValueError as e:
    print(f'Ошибка: {e} Повторите ввод!')
```

Условие

Задача «Ферзи»
Условие
Известно, что на доске 8×8 можно расставить 8 ферзей так, чтобы они не били друг друга.
Вам дана расстановка 8 ферзей на доске, определите, есть ли среди них пара бьющих друг
друга.
Программа получает на вход восемь пар чисел, каждое число от 1 до 8 - координаты 8
ферзей. Если ферзи не бьют друг друга, выведите слово NO , иначе выведите YES .

```python
try:
    def is_atacking(q1, q2):
        return q1[0] == q2[0] or q1[1] == q2[1] or abs(q1[0] - q2[0]) == abs(q1[1] - q2[1])


    list_quens = []
    for _ in range(8):
        x, y = map(int, input('x, y: ').split())
        list_quens.append((x, y))

    found = False
    for i in range(len(list_quens)):
        for j in range(i + 1, len(list_quens)):
            if is_atacking(list_quens[i], list_quens[j]):
                # print('YES')
                # exit()
                found = True
                break
        if found:
            break
    if found:
        print('YES')
    else:
        print('NO')
except ValueError as e:
    print(f'Ошибка: {e} Повторите ввод!')
```

Окунёмся в суровый и беспощадный мир очереди на приём к врачу в поликлинике.
Напишите программу-медсестру, которая поможет врачу определить, кто следующий
проходит к нему в кабинет.
Изначально в очереди никого нет. Далее в каком-то порядке происходят события трёх
типов:
 фраза «Кто последний? Я - <фамилия>.» означает, что пришёл новый пациент и
встал в конец очереди;
 фраза «Я только спросить! Я - <фамилия>.» означает, что пришёл новый пациент и
встал в начало очереди;
 фраза «Следующий!» означает, что доктор готов принять первого пациента в
очереди.
В последнем случае нужно вывести «Заходит <фамилия>!» или «В очереди никого нет.»
На психологическую достоверность данная модель очереди, конечно, не претендует: в
жизни такое нахальство с «я только спросить» прощается далеко не всегда.

```python
try:
    queue = []
    
    while True:
        # Ввод команды
        command = input("Введите команду: ")
    
        if command.startswith("Кто последний? Я - "):
            # Пациент встаёт в конец очереди
            surname = command.split("Я - ")[1].strip(".")
            queue.append(surname)
            print(f"{surname} добавлен в конец очереди.")
    
        elif command.startswith("Я только спросить! Я - "):
            # Пациент встаёт в начало очереди
            surname = command.split("Я - ")[1].strip(".")
            queue.insert(0, surname)
            print(f"{surname} добавлен в начало очереди.")
    
        elif command == "Следующий!":
            # Врач вызывает следующего пациента
            if queue:
                next_patient = queue.pop(0)
                print(f"Заходит {next_patient}!")
            else:
                print("В очереди никого нет.")
    
        else:
            print("Неизвестная команда. Попробуйте снова.")
except ValueError as e:
    print(f'Ошибка: {e} Повторите ввод!')
```
